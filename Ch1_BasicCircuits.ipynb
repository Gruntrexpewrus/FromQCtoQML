```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import *
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

Cnot

```python
a = QuantumRegister(1, 'a')
b = QuantumRegister(1, 'b')
qc = QuantumCircuit(a,b)
qc.cnot(0, 1)
save = qc.draw()
```

```python
save.savefig('Cnot.png')
```

Hadamard Circuit

```python
qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)
sav = qc.draw('mpl')
sav.savefig('Hcirc')
sav
```

```python
backend_sim = Aer.get_backend('aer_simulator')
result = execute(qc, backend_sim, shots = 10000).result()
counts = result.get_counts()
save = plot_histogram(counts)
save.savefig('histoH.png')
save
```

The Entangling circuit

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cnot(0, 1)
qc.measure_all()
sav = qc.draw('mpl')
sav.savefig('Entanglingcirc')
sav
```

```python
backend_sim = Aer.get_backend('aer_simulator')
result = execute(qc, backend_sim, shots = 10000).result()
counts = result.get_counts()
save = plot_histogram(counts)
save.savefig('histoEntangl.png')
save
```
