```python
pls = QuantumCircuit(1)
pls.h(0)
pls.save_statevector()
pls.measure_all()
pls.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
res = execute(pls, sim, shots = 100000).result()
```

```python
out_state = res.get_statevector()
print(out_state)
```

```python
counts = res.get_counts()
o = plot_histogram(counts, title = "Superposition between 0 and 1")
o.savefig("histH.png")
```
